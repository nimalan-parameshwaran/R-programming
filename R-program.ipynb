```r
# Sample data with NA values
values <- c(3, 5, 7, NA, 4, 9)

# Calculating mean without removing NA values
result_with_na <- mean(values)
print(result_with_na)  # Output will be NA

# Calculating mean after removing NA values
result_without_na <- mean(values, na.rm = TRUE)
print(result_without_na)  # Output will be the mean of non-NA values
```

```r
# Sample data with NA values
values <- c(3, 5, 7, NA, 4, 9)

# Calculating mean without removing NA values
result_with_na <-median(values)
print(result_with_na)  # Output will be NA

# Calculating mean after removing NA values
result_without_na <- median(values, na.rm = TRUE)
print(result_without_na)  # Output will be the mean of non-NA values
```

# **Natural Join or Inner Join**


```r
df1 = data.frame(StudentId = c(101:106), Product = c("Hindi", "English","Maths", "Science", "Political Science", "Physics"))
df1
```

```r
df2 = data.frame(StudentId = c(102, 104, 106,107, 108),State = c("Mangalore", "Mysore","Pune", "Dehradun", "Delhi"))
df2
```

```r
df = merge(x = df1, y = df2)
df
```

```r
df = merge(x = df1, y = df2, by = "StudentId")
df
```

# **Left Outer Join**

```r
df = merge(x = df1, y = df2, by = "StudentId",all.x = TRUE)
df
```

# **Right Outer Join**

```r
df = merge(x = df1, y = df2, by = "StudentId",all.y = TRUE)
df
```

# **Full Outer Join**

```r
df = merge(x = df1, y = df2, by = "StudentId",all = TRUE)
df
```

# **Cross Join**

```r
# R program to illustrate
# Joining of dataframes

df = merge(x = df1, y = df2, by = NULL)
df
```
